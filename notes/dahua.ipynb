```python
import requests
from requests.auth import HTTPBasicAuth, HTTPDigestAuth
import subprocess
import cv2
import numpy as np

command = "curl  --digest -u admin:1234567a \"http://172.16.115.203/cgi-bin/snapshot.cgi\" --output -"
def exec_shell(cmd):
    try:
        output = subprocess.check_output(cmd, shell=True)
        return output
    except subprocess.CalledProcessError as e:
        print(e)
        return None
r = exec_shell(command)

raw_jpeg = np.frombuffer(r, np.uint8)
bgr_image = cv2.imdecode(raw_jpeg, cv2.IMREAD_COLOR)
```
