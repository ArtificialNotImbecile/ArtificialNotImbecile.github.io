# A(A1 and A2 is the same) is the desire behaviour, but B(B1 and B2 is the same) is not. Because the method resolution order is first apply decorator, then execute function.
since
```
@decorator1
@decorator2
def fun():
    pass
# same as
decorator1(decorator2(fun))
```
is not the same as
```
decorator = decorator1(decorator2)
@decorator
def fun():
    pass
# same as
decorator1(decorator2)(fun)
```
